# 4.3 内存优化

端侧设备内存有限，内存优化是部署的关键。核心技术包括激活重计算、内存复用规划和权重按需加载。

## 什么是内存优化？

内存优化是在保证模型输出数学等价的前提下，通过算法和系统层面的技术降低模型运行时的峰值内存占用。核心方法包括：
- **激活重计算/梯度检查点**：前向时不保存中间激活，反向时重新计算，以计算换内存
- **内存复用规划**：让生命周期不重叠的张量共享同一块内存空间
- **权重按需加载**：逐层从慢速存储加载权重，不将全部权重常驻内存

## 为什么用内存优化？

1. **峰值内存决定能否部署**：模型能否在设备上运行取决于$M_{\text{peak}} < M_{\text{available}}$
2. **内存-计算权衡**：内存优化通常以增加计算量为代价，但在内存受限场景下这是唯一选择
3. **内存带宽也是瓶颈**：端侧设备内存带宽远低于云端GPU

## 内存优化的数学原理

**梯度检查点**：$L$层网络，每$k$层设一个检查点：
$$M_{\text{checkpoint}} = \frac{L}{k} |A_{\text{ckpt}}| + k |A_{\text{layer}}|$$
当$k = \sqrt{L}$时峰值内存最小为$O(\sqrt{L} \cdot |A|)$，相比标准模式$O(L \cdot |A|)$节省$O(\sqrt{L})$倍。

**内存复用**：将张量建模为区间$[b_i, d_i)$，复用等价于区间图着色问题，贪心策略可节省$30\%\!-\!60\%$峰值内存。

**权重流式加载**：$M_{\text{streaming}} = w_{\text{single layer}} + |A| \ll W + |A| = M_{\text{standard}}$，代价是每层需额外I/O延迟。

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import time
import gc

torch.manual_seed(42)
np.random.seed(42)

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")

def benchmark(fn, *args, warmup=10, runs=100, **kwargs):
    """通用benchmark函数: warmup + 多次运行取平均"""
    for _ in range(warmup):
        fn(*args, **kwargs)
    if torch.cuda.is_available():
        torch.cuda.synchronize()
    start = time.perf_counter()
    for _ in range(runs):
        fn(*args, **kwargs)
    if torch.cuda.is_available():
        torch.cuda.synchronize()
    elapsed = (time.perf_counter() - start) / runs * 1000
    return elapsed
````

### 激活重计算（Activation Recomputation / Gradient Checkpointing）

#### 什么是梯度检查点？

前向传播时仅保存部分关键层（检查点）的输入激活，其余层的中间激活不保存。反向传播需要某层激活时，从最近的检查点重新执行前向计算来恢复。

#### 为什么梯度检查点有效？

1. **内存大幅降低**：标准模式$M = O(L \cdot |A|)$，检查点模式$M = O(\sqrt{L} \cdot |A|)$
2. **额外计算可控**：最多增加约33%的前向计算
3. **训练必需**：大模型训练时内存瓶颈往往是激活值而非权重

#### 梯度检查点的数学原理

设检查点间隔为$k$，$L$层网络：
- 保存的检查点数量：$\lceil L/k \rceil$
- 每段重计算的最大层数：$k$
- 峰值内存：$M(k) = \frac{L}{k}|A| + k|A| = |A| \cdot (\frac{L}{k} + k)$

对$f(k) = \frac{L}{k} + k$求导令其为零：$f'(k) = -\frac{L}{k^2} + 1 = 0 \Rightarrow k^* = \sqrt{L}$

此时峰值内存最小为$2\sqrt{L}|A|$，相比标准模式$L|A|$节省$O(\sqrt{L})$倍。

````python
class TransformerLayer(nn.Module):
    """标准Transformer层: Pre-LN + Self-Attention + FFN"""
    def __init__(self, dim: int, n_heads: int = 4, ffn_dim: int = None):
        super().__init__()
        ffn_dim = ffn_dim or dim * 4
        self.attn = nn.MultiheadAttention(dim, n_heads, batch_first=True)
        self.ffn = nn.Sequential(nn.Linear(dim, ffn_dim), nn.GELU(), nn.Linear(ffn_dim, dim))
        self.ln1 = nn.LayerNorm(dim)
        self.ln2 = nn.LayerNorm(dim)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        h = self.ln1(x)
        h, _ = self.attn(h, h, h)
        x = x + h
        x = x + self.ffn(self.ln2(x))
        return x

class StandardTransformer(nn.Module):
    """标准Transformer: 保存所有中间激活"""
    def __init__(self, dim: int, n_layers: int = 6, n_heads: int = 4):
        super().__init__()
        self.layers = nn.ModuleList([TransformerLayer(dim, n_heads) for _ in range(n_layers)])

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        for layer in self.layers:
            x = layer(x)
        return x

class CheckpointedTransformer(nn.Module):
    """带梯度检查点的Transformer: 仅保存边界激活"""
    def __init__(self, dim: int, n_layers: int = 6, n_heads: int = 4, checkpoint_every: int = 2):
        super().__init__()
        self.layers = nn.ModuleList([TransformerLayer(dim, n_heads) for _ in range(n_layers)])
        self.checkpoint_every = checkpoint_every

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        for i, layer in enumerate(self.layers):
            if i % self.checkpoint_every == 0 and i < len(self.layers) - 1:
                x = torch.utils.checkpoint.checkpoint(layer, x, use_reentrant=False)
            else:
                x = layer(x)
        return x

dim, n_layers, n_heads = 128, 6, 4
x = torch.randn(4, 32, dim, requires_grad=True)

model_std = StandardTransformer(dim, n_layers, n_heads)
model_ckpt = CheckpointedTransformer(dim, n_layers, n_heads, checkpoint_every=2)

for p1, p2 in zip(model_std.parameters(), model_ckpt.parameters()):
    p2.data.copy_(p1.data)

if torch.cuda.is_available():
    torch.cuda.reset_peak_memory_stats()

out_std = model_std(x)
loss_std = out_std.sum()
loss_std.backward()

out_ckpt = model_ckpt(x)
loss_ckpt = out_ckpt.sum()
loss_ckpt.backward()

diff = (out_std.detach() - out_ckpt.detach()).norm() / out_std.detach().norm() * 100

activation_mem_per_layer = 4 * 32 * dim * 4
total_activations = n_layers * activation_mem_per_layer
checkpoint_savings = (n_layers - n_layers // 2) * activation_mem_per_layer

print(f"=== 激活重计算 (Gradient Checkpointing) ===")
print(f"层数: {n_layers}, 每{2}层设一个检查点")
print(f"输出差异: {diff:.6f}% (数学等价)")
print(f"\n内存分析:")
print(f"  每层激活大小: {activation_mem_per_layer/1024:.2f} KB")
print(f"  标准模式总激活: {total_activations/1024:.2f} KB")
print(f"  检查点模式激活: {(total_activations - checkpoint_savings)/1024:.2f} KB")
print(f"  节省: {checkpoint_savings/total_activations*100:.0f}%")
print(f"\n代价: 需要额外~33%的前向计算(重计算)")
````

### 内存复用规划（Memory Planning）

#### 什么是内存复用规划？

分析计算图中所有张量的生命周期（出生步-死亡步），让生命周期不重叠的张量共享同一块物理内存，从而降低峰值内存占用。

#### 为什么需要内存规划？

1. **朴素策略浪费严重**：为每个张量独立分配内存，峰值内存等于所有活跃张量大小之和
2. **生命周期不重叠**：很多张量生命周期不重叠，一个张量死亡后其内存可被新张量复用
3. **端侧内存有限**：端侧设备每MB内存都很宝贵

#### 内存复用的数学原理

将每个张量$T_i$建模为区间$[b_i, d_i)$：
- 两个张量$T_i, T_j$可共享内存$\iff [b_i, d_i) \cap [b_j, d_j) = \varnothing$
- 峰值内存：$M_{\text{peak}} = \max_t \sum_{i: t \in [b_i, d_i)} |T_i|$
- 复用后峰值内存：$M_{\text{planned}} = M_{\text{peak}} - \text{Reuse Savings}$

最优内存规划是NP-hard问题，但贪心策略（First-Fit/Best-Fit）可节省30%-60%峰值内存。

````python
class MemoryPlanner:
    """静态内存规划器：分析张量生命周期，复用不再使用的内存"""
    def __init__(self):
        self.tensors = {}

    def allocate(self, name: str, size_bytes: int, birth_step: int, death_step: int):
        self.tensors[name] = {
            'size': size_bytes,
            'birth': birth_step,
            'death': death_step
        }

    def plan_naive(self) -> int:
        """朴素规划: 每个张量独立分配，返回峰值内存"""
        peak = 0
        max_step = max(t['death'] for t in self.tensors.values())
        for step in range(max_step + 1):
            active_mem = sum(
                t['size'] for t in self.tensors.values()
                if t['birth'] <= step < t['death']
            )
            peak = max(peak, active_mem)
        return peak

    def plan_reuse(self) -> int:
        """复用规划: 生命周期不重叠的张量共享内存 (First-Fit策略)"""
        events = []
        for name, info in self.tensors.items():
            events.append((info['birth'], 'alloc', name, info['size']))
            events.append((info['death'], 'free', name, info['size']))
        events.sort(key=lambda x: (x[0], x[1] == 'alloc'))

        free_blocks = []
        next_addr = 0
        peak_memory = 0
        active = {}

        for step, action, name, size in events:
            if action == 'alloc':
                reused = False
                for i, (block_size, block_addr) in enumerate(free_blocks):
                    if block_size >= size:
                        addr = block_addr
                        remaining = block_size - size
                        free_blocks.pop(i)
                        if remaining > 0:
                            free_blocks.append((remaining, addr + size))
                        active[name] = (addr, size)
                        reused = True
                        break
                if not reused:
                    addr = next_addr
                    next_addr += size
                    active[name] = (addr, size)

                current_usage = next_addr - sum(bs for bs, _ in free_blocks)
                peak_memory = max(peak_memory, next_addr)
            elif action == 'free':
                if name in active:
                    addr, sz = active.pop(name)
                    free_blocks.append((sz, addr))

        return peak_memory

planner = MemoryPlanner()

layer_mem = 4 * 32 * 128 * 4
for i in range(6):
    planner.allocate(f'attn_out_{i}', layer_mem, i*2, i*2+1)
    planner.allocate(f'ffn_out_{i}', layer_mem, i*2+1, i*2+2)

naive_mem = planner.plan_naive()
reuse_mem = planner.plan_reuse()

print(f"=== 内存复用规划 ===")
print(f"张量数量: {len(planner.tensors)}")
print(f"朴素分配峰值内存: {naive_mem/1024:.2f} KB")
print(f"复用规划峰值内存: {reuse_mem/1024:.2f} KB")
print(f"节省: {(1-reuse_mem/naive_mem)*100:.0f}%")
print(f"\n内存复用核心: 生命周期不重叠的张量共享同一块内存")
print(f"\n产业界内存规划:")
print(f"1. TVM: 编译期静态内存规划，零运行时开销")
print(f"2. MLC-LLM: 基于TVM的端侧内存规划")
print(f"3. PyTorch: torch.compile后端自动内存优化")
print(f"4. ONNX Runtime: 运行时内存池复用")
````

### 权重按需加载（Weight Streaming / Offloading）

#### 什么是权重按需加载？

当模型总权重大小超过可用内存时，不将全部权重常驻内存，而是逐层从慢速存储（磁盘/SSD/CPU内存）加载到快速内存，计算完成后释放，再加载下一层权重。

#### 为什么需要权重按需加载？

1. **大模型超出内存**：LLaMA-7B INT4量化后仍需约3.5GB，手机可用内存可能仅2-4GB
2. **标准推理需全部权重常驻**：$M_{\text{standard}} = W + |A|$
3. **流式加载仅需单层权重**：$M_{\text{streaming}} = w_{\text{layer}} + |A|$

#### 权重按需加载的数学原理

- 内存节省：$\frac{M_{\text{streaming}}}{M_{\text{standard}}} \approx \frac{1}{L}$（$L$为层数）
- 延迟增加：每层需额外$t_{\text{load}} = \frac{w}{\text{BW}_{\text{storage}}}$的I/O时间
- 异步预取优化：计算第$l$层时同时加载第$l+1$层权重：
$$T_{\text{prefetch}} = T_{\text{compute}} + t_{\text{load}} + (L-1) \cdot \max(0, t_{\text{load}} - t_{\text{compute}}^{\text{layer}})$$

````python
class StreamingInference:
    """真实的权重流式加载推理
    逐层加载权重 → 执行推理 → 释放权重 → 加载下一层
    适用于模型超过可用内存的场景"""
    def __init__(self, dim: int = 256, n_layers: int = 8, n_heads: int = 4, ffn_dim: int = 1024):
        self.dim = dim
        self.n_layers = n_layers
        self.layer_weights = {}
        for i in range(n_layers):
            layer = TransformerLayer(dim=dim, n_heads=n_heads, ffn_dim=ffn_dim)
            self.layer_weights[i] = {k: v.clone() for k, v in layer.state_dict().items()}
            del layer

    def inference(self, x: torch.Tensor, memory_budget_layers: int = 1) -> tuple:
        """流式推理: 每次仅加载memory_budget_layers层权重到内存"""
        loaded_layers = {}
        peak_layer_count = 0
        current_layer_count = 0

        for i in range(self.n_layers):
            if i not in loaded_layers:
                while current_layer_count >= memory_budget_layers and loaded_layers:
                    oldest = min(loaded_layers.keys())
                    del loaded_layers[oldest]
                    current_layer_count -= 1
                layer = TransformerLayer(dim=self.dim)
                layer.load_state_dict(self.layer_weights[i])
                layer.eval()
                loaded_layers[i] = layer
                current_layer_count += 1

            with torch.no_grad():
                x = loaded_layers[i](x)
            peak_layer_count = max(peak_layer_count, current_layer_count)

        return x, peak_layer_count

streamer = StreamingInference(dim=256, n_layers=8, n_heads=4, ffn_dim=1024)
x = torch.randn(1, 32, 256)

with torch.no_grad():
    out_full, peak_full = streamer.inference(x, memory_budget_layers=8)
    out_stream, peak_stream = streamer.inference(x, memory_budget_layers=1)

diff = (out_full - out_stream).abs().max().item()

single_layer_mem = sum(p.numel() * p.element_size() for p in TransformerLayer(256, 4, 1024).parameters()) / 1024 / 1024
full_model_mem = single_layer_mem * streamer.n_layers

print(f"=== 权重流式加载推理 ===")
print(f"模型: {streamer.n_layers}层Transformer")
print(f"单层内存: {single_layer_mem:.2f} MB")
print(f"全量加载: {full_model_mem:.2f} MB")
print(f"\n{'模式':<25} {'峰值层数':<12} {'峰值内存(MB)':<15} {'节省':<10}")
print("-" * 62)
print(f"{'全量加载':<25} {peak_full:<12} {full_model_mem:<15.2f} {'-':<10}")
print(f"{'流式加载(budget=1)':<25} {peak_stream:<12} {single_layer_mem:<15.2f} {(1-single_layer_mem/full_model_mem)*100:.0f}%")
print(f"\n输出一致性: 最大差异 = {diff:.8f}")

print(f"\n产业界权重流式加载实践:")
print(f"1. llama.cpp: mmap加载 + 逐层推理 (端侧CPU首选)")
print(f"2. MLC-LLM: 编译期权重分块 + 按需加载")
print(f"3. 权重预取: 在计算当前层时异步加载下一层")
print(f"4. MoE场景: 仅加载被路由选中的专家权重")
print(f"5. 代价: 增加IO延迟，适合吞吐不敏感、内存受限场景")
````

### 产业级实战：真实梯度检查点内存测量

使用 PyTorch 原生 `torch.utils.checkpoint` 在真实 Transformer 模型上测量内存节省效果。

#### 关键API

- `torch.utils.checkpoint.checkpoint(fn, *args)`：将函数`fn`标记为检查点边界
- 前向传播时不保存`fn`内部的中间激活，仅保存输入
- 反向传播时自动重新执行`fn`的前向计算来恢复激活

#### 内存测量方法

```python
torch.cuda.reset_peak_memory_stats()
# ... forward + backward ...
peak_mem = torch.cuda.max_memory_allocated()
```

#### 预期结果

- 标准模式：$M \propto L$（与层数线性增长）
- 检查点模式：$M \propto \sqrt{L}$（与层数平方根增长）

````python
from torch.utils.checkpoint import checkpoint

class DeepTransformer(nn.Module):
    def __init__(self, dim: int = 256, depth: int = 12, n_heads: int = 4, use_checkpoint: bool = False):
        super().__init__()
        self.use_checkpoint = use_checkpoint
        self.layers = nn.ModuleList([TransformerLayer(dim, n_heads) for _ in range(depth)])
        self.embed = nn.Linear(dim, dim)
        self.head = nn.Linear(dim, dim)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.embed(x)
        for layer in self.layers:
            if self.use_checkpoint and self.training:
                x = checkpoint(layer, x, use_reentrant=False)
            else:
                x = layer(x)
        return self.head(x)

def measure_training_memory(model: nn.Module, batch_size: int = 4, seq_len: int = 64, dim: int = 256) -> float:
    """测量训练峰值内存（MB）"""
    model.train()
    device = next(model.parameters()).device
    x = torch.randn(batch_size, seq_len, dim, device=device)
    target = torch.randn(batch_size, seq_len, dim, device=device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)

    if device.type == 'cuda':
        torch.cuda.reset_peak_memory_stats(device)
        torch.cuda.empty_cache()

    output = model(x)
    loss = F.mse_loss(output, target)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if device.type == 'cuda':
        peak_mem = torch.cuda.max_memory_allocated(device) / 1024 / 1024
    else:
        param_mem = sum(p.numel() * p.element_size() for p in model.parameters()) / 1024 / 1024
        activation_mem = batch_size * seq_len * dim * 4 * 4 / 1024 / 1024
        peak_mem = param_mem * 2 + activation_mem * 3
    return max(peak_mem, 0.01)

dim, depth = 256, 12
model_no_ckpt = DeepTransformer(dim=dim, depth=depth, use_checkpoint=False)
model_with_ckpt = DeepTransformer(dim=dim, depth=depth, use_checkpoint=True)
model_with_ckpt.load_state_dict(model_no_ckpt.state_dict())

mem_no_ckpt = measure_training_memory(model_no_ckpt)
mem_with_ckpt = measure_training_memory(model_with_ckpt)

param_count = sum(p.numel() for p in model_no_ckpt.parameters())

print(f"=== 梯度检查点内存对比 ===")
print(f"模型: {depth}层Transformer, {param_count:,}参数")
print(f"{'配置':<25} {'峰值内存(MB)':<15} {'节省':<10}")
print("-" * 50)
print(f"{'无检查点':<25} {mem_no_ckpt:<15.1f} {'-':<10}")
print(f"{'梯度检查点':<25} {mem_with_ckpt:<15.1f} {(1-mem_with_ckpt/mem_no_ckpt)*100:.0f}%")

print(f"\n产业界梯度检查点使用指南:")
print(f"1. API: torch.utils.checkpoint.checkpoint(layer, x, use_reentrant=False)")
print(f"2. 原理: 前向不保存中间激活，反向时重计算")
print(f"3. 效果: 约30%额外计算，节省60%+峰值内存")
print(f"4. 适用: 端侧训练(内存受限)、大模型微调(QLoRA)")
print(f"5. 注意: 随机性操作需用use_reentrant=False确保正确性")

del model_no_ckpt, model_with_ckpt
gc.collect()
````

### 选择性检查点（Selective Checkpointing）

#### 什么是选择性检查点？

不是对所有层都做检查点，而是只对内存占用大的层（如Attention层）做检查点，对内存占用小的层（如LayerNorm）保留激活。

#### 为什么需要选择性检查点？

1. **精细的内存-计算权衡**：全量检查点可能过度增加计算，选择性检查点可精确控制
2. **Attention层占用大**：Attention的QKV中间结果占用$O(n^2 \cdot d)$内存，是检查点的首选目标
3. **LayerNorm占用小**：LayerNorm中间结果很小，不值得重计算

#### 选择性检查点策略

| 策略 | 检查点范围 | 内存节省 | 额外计算 |
|------|-----------|---------|----------|
| 无检查点 | 无 | 0% | 0% |
| 全量检查点 | 每层 | ~60% | ~33% |
| 选择性检查点 | 仅Attention | ~40% | ~15% |
| 选择性检查点 | 仅FFN | ~30% | ~10% |

````python
class SelectiveCheckpointTransformer(nn.Module):
    """选择性检查点: 仅对Attention层做检查点"""
    def __init__(self, dim: int, n_layers: int = 6, n_heads: int = 4, checkpoint_attn: bool = True):
        super().__init__()
        self.layers = nn.ModuleList([TransformerLayer(dim, n_heads) for _ in range(n_layers)])
        self.checkpoint_attn = checkpoint_attn

    def _attn_forward(self, x: torch.Tensor, layer: TransformerLayer) -> torch.Tensor:
        """单独的Attention前向（用于检查点）"""
        h = layer.ln1(x)
        h, _ = layer.attn(h, h, h)
        return x + h

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        for layer in self.layers:
            if self.checkpoint_attn and self.training:
                x = checkpoint(self._attn_forward, x, layer, use_reentrant=False)
            else:
                h = layer.ln1(x)
                h, _ = layer.attn(h, h, h)
                x = x + h
            x = x + layer.ffn(layer.ln2(x))
        return x

dim, n_layers, n_heads = 128, 6, 4
model_no_ckpt = StandardTransformer(dim, n_layers, n_heads)
model_full_ckpt = CheckpointedTransformer(dim, n_layers, n_heads, checkpoint_every=1)
model_sel_ckpt = SelectiveCheckpointTransformer(dim, n_layers, n_heads, checkpoint_attn=True)

for p1, p2, p3 in zip(model_no_ckpt.parameters(), model_full_ckpt.parameters(), model_sel_ckpt.parameters()):
    p2.data.copy_(p1.data)
    p3.data.copy_(p1.data)

x = torch.randn(4, 32, dim, requires_grad=True)

models = {
    '无检查点': model_no_ckpt,
    '全量检查点': model_full_ckpt,
    '选择性检查点(仅Attn)': model_sel_ckpt,
}

print(f"=== 选择性检查点对比 ===")
print(f"模型: {n_layers}层Transformer, dim={dim}")
print(f"\n{'策略':<25} {'输出范数':<12} {'与无检查点差异':<15}")
print("-" * 52)

ref_out = None
for name, model in models.items():
    model.train()
    x_input = torch.randn(4, 32, dim, requires_grad=True)
    out = model(x_input)
    loss = out.sum()
    loss.backward()

    if ref_out is None:
        ref_out = out.detach()
        diff_str = '-'
    else:
        diff = (ref_out - out.detach()).norm() / ref_out.norm() * 100
        diff_str = f'{diff:.4f}%'

    print(f"{name:<25} {out.detach().norm():<12.4f} {diff_str:<15}")

print(f"\n选择性检查点优势:")
print(f"1. 仅重计算内存占用大的Attention层")
print(f"2. FFN和LayerNorm的激活保留，不增加额外计算")
print(f"3. 内存节省介于无检查点和全量检查点之间")
print(f"4. 适合需要精细控制内存-计算权衡的场景")
````

### KV Cache 内存优化：LLM推理的核心瓶颈

#### 什么是KV Cache？

自回归LLM推理时，每生成一个新token都需要访问之前所有token的Key和Value。KV Cache将已计算的KV存储下来避免重复计算，但内存占用随序列长度线性增长。

#### KV Cache内存公式

$$M_{\text{KV}} = 2 \cdot n_{\text{layers}} \cdot n_{\text{kv\_heads}} \cdot d_{\text{head}} \cdot L_{\text{seq}} \cdot b_{\text{bytes}}$$

其中$b_{\text{bytes}}$为每个元素的字节数（FP16=2, INT8=1, INT4=0.5）。

#### KV Cache优化策略

1. **GQA/MQA**：减少KV头数，$n_{\text{kv\_heads}} \ll n_{\text{heads}}$
2. **KV Cache量化**：FP16→INT8→INT4，内存减半/四分之一
3. **PagedAttention**：非连续内存管理，消除碎片
4. **滑动窗口**：只保留最近$W$个token的KV，$M_{\text{KV}} \propto W$
5. **Token淘汰**：注意力分数低的token的KV可被丢弃

````python
def compute_kv_cache_memory(n_layers: int, n_kv_heads: int, head_dim: int, seq_len: int, dtype_bytes: float = 2) -> float:
    """计算KV Cache内存占用（字节）"""
    return 2 * n_layers * n_kv_heads * head_dim * seq_len * dtype_bytes

configs = [
    ('LLaMA-2-7B MHA', 32, 32, 128, 2),
    ('LLaMA-2-7B GQA', 32, 8, 128, 2),
    ('LLaMA-2-7B GQA+INT8', 32, 8, 128, 1),
    ('LLaMA-2-7B GQA+INT4', 32, 8, 128, 0.5),
]

print(f"=== KV Cache 内存分析 (seq_len=2048) ===")
print(f"{'配置':<25} {'KV Heads':<10} {'精度':<8} {'KV Cache(MB)':<15}")
print("-" * 58)
for name, nl, nkv, hd, db in configs:
    mem = compute_kv_cache_memory(nl, nkv, hd, 2048, db) / 1024 / 1024
    dtype_str = {2: 'FP16', 1: 'INT8', 0.5: 'INT4'}[db]
    print(f"{name:<25} {nkv:<10} {dtype_str:<8} {mem:<15.1f}")

print(f"\n=== PagedAttention 原理 ===")
print(f"传统KV Cache: 预分配最大seq_len的连续内存 → 大量碎片")
print(f"PagedAttention: 将KV Cache分为固定大小的page，按需分配")
print(f"  - Page大小: 16个token的KV (典型值)")
print(f"  - 物理页可以不连续，通过页表映射")
print(f"  - 共享前缀: 不同请求可共享相同前缀的KV页（如system prompt）")
print(f"  - 内存节省: 50-80% (vLLM实测)")
````

````python
class SimplePagedKVCache:
    """简化PagedAttention KV Cache管理器"""
    def __init__(self, page_size: int = 16, n_pages: int = 256, head_dim: int = 128, n_kv_heads: int = 8, n_layers: int = 4):
        self.page_size = page_size
        self.n_pages = n_pages
        self.page_table = {}
        self.free_pages = list(range(n_pages))
        self.page_size_bytes = 2 * n_layers * page_size * n_kv_heads * head_dim * 2

    def allocate(self, request_id: str, n_tokens: int) -> bool:
        n_pages_needed = (n_tokens + self.page_size - 1) // self.page_size
        if len(self.free_pages) < n_pages_needed:
            return False
        pages = self.free_pages[:n_pages_needed]
        self.free_pages = self.free_pages[n_pages_needed:]
        self.page_table[request_id] = pages
        return True

    def free(self, request_id: str):
        if request_id in self.page_table:
            self.free_pages.extend(self.page_table.pop(request_id))

    @property
    def total_memory_mb(self) -> float:
        return self.n_pages * self.page_size_bytes / 1024 / 1024

    @property
    def used_pages(self) -> int:
        return self.n_pages - len(self.free_pages)

paged_cache = SimplePagedKVCache(n_pages=256)

print(f"=== Paged KV Cache 实例 ===")
print(f"总页数: {paged_cache.n_pages}")
print(f"每页token数: {paged_cache.page_size}")
print(f"总内存池: {paged_cache.total_memory_mb:.2f} MB")

requests = [(f'req_{i}', np.random.randint(64, 256)) for i in range(5)]
for rid, ntok in requests:
    success = paged_cache.allocate(rid, ntok)
    used_pct = paged_cache.used_pages / paged_cache.n_pages * 100
    print(f"  {rid}: {ntok} tokens → {'成功' if success else '失败(内存不足)'} (使用率{used_pct:.0f}%)")

paged_cache.free('req_0')
print(f"释放req_0后可用页: {len(paged_cache.free_pages)}")

print(f"\n产业界KV Cache优化方案:")
print(f"1. vLLM: PagedAttention + 连续批处理 (GPU服务端)")
print(f"2. llama.cpp: 滑动窗口 + KV量化 (端侧CPU)")
print(f"3. FlashInfer: 批量解码优化 (GPU)")
print(f"4. HuggingFace: KV Cache量化 + 共享前缀")
````

### 产业级实战：CPU Offloading 大模型推理

#### 什么是CPU Offloading？

当GPU显存不足时，将部分模型权重或KV Cache卸载到CPU内存，需要时再传输到GPU。这是权重按需加载在GPU场景的具体实现。

#### 为什么需要CPU Offloading？

1. **GPU显存有限**：消费级GPU通常8-24GB显存，LLaMA-70B FP16需140GB
2. **CPU内存充裕**：系统内存通常64-128GB，可容纳完整模型
3. **延迟换内存**：PCIe带宽约32GB/s，额外延迟可接受

#### Offloading策略

| 策略 | GPU内存 | CPU内存 | 延迟 |
|------|---------|---------|------|
| 全GPU | W + A | 0 | 最低 |
| 权重Offload | w_layer + A | W | 中等 |
| KV Offload | W | KV Cache | 较高 |
| 混合Offload | 部分W + A | 部分W + KV | 可调 |

````python
class CPUOffloadInference:
    """模拟CPU Offloading推理: 权重存储在CPU，逐层传输到设备执行"""
    def __init__(self, dim: int = 256, n_layers: int = 6, n_heads: int = 4, ffn_dim: int = 1024):
        self.dim = dim
        self.n_layers = n_layers
        self.cpu_weights = {}
        for i in range(n_layers):
            layer = TransformerLayer(dim=dim, n_heads=n_heads, ffn_dim=ffn_dim)
            self.cpu_weights[i] = {k: v.cpu().clone() for k, v in layer.state_dict().items()}
            del layer

    def inference(self, x: torch.Tensor, device: str = 'cpu') -> tuple:
        """逐层加载权重到设备执行推理"""
        peak_device_mem = 0
        total_transfer_bytes = 0

        for i in range(self.n_layers):
            layer = TransformerLayer(dim=self.dim)
            layer.load_state_dict(self.cpu_weights[i])
            layer.eval()
            layer.to(device)

            layer_mem = sum(p.numel() * p.element_size() for p in layer.parameters())
            total_transfer_bytes += layer_mem
            peak_device_mem = max(peak_device_mem, layer_mem + x.numel() * x.element_size())

            with torch.no_grad():
                x = layer(x)

            layer.to('cpu')
            del layer

        return x, peak_device_mem, total_transfer_bytes

offloader = CPUOffloadInference(dim=256, n_layers=6, n_heads=4, ffn_dim=1024)
x = torch.randn(1, 32, 256)

with torch.no_grad():
    out, peak_mem, transfer_bytes = offloader.inference(x)

full_model_mem = sum(
    sum(v.numel() * v.element_size() for v in layer_weights.values())
    for layer_weights in offloader.cpu_weights.values()
)

print(f"=== CPU Offloading 推理 ===")
print(f"模型: {offloader.n_layers}层Transformer")
print(f"全量加载设备内存: {full_model_mem/1024/1024:.2f} MB")
print(f"Offload峰值设备内存: {peak_mem/1024/1024:.2f} MB")
print(f"内存节省: {(1-peak_mem/full_model_mem)*100:.0f}%")
print(f"数据传输量: {transfer_bytes/1024/1024:.2f} MB")

print(f"\n产业界CPU Offloading方案:")
print(f"1. DeepSpeed ZeRO-Offload: 优化器状态+梯度卸载到CPU")
print(f"2. llama.cpp: CPU推理 + 可选GPU加速层")
print(f"3. HuggingFace Accelerate: device_map='auto' 自动分配")
print(f"4. AirLLM: 逐层加载 + 4bit量化 + CPU offload")
print(f"5. 关键优化: 异步预取下一层权重，隐藏传输延迟")
````

### 本章总结

#### 内存优化技术全景

| 技术 | 原理 | 内存节省 | 额外代价 | 适用场景 |
|------|------|---------|---------|----------|
| 梯度检查点 | 重计算中间激活 | ~60% | ~33%额外计算 | 训练/微调 |
| 选择性检查点 | 仅重计算大层 | ~40% | ~15%额外计算 | 精细控制 |
| 内存复用规划 | 生命周期不重叠共享 | 30-60% | 编译期开销 | 所有场景 |
| 权重流式加载 | 逐层加载权重 | ~1/L | I/O延迟 | 内存受限推理 |
| CPU Offloading | 权重卸载到CPU | 显著 | PCIe传输 | GPU显存不足 |
| KV Cache量化 | 低精度存储KV | 50-75% | 精度损失 | LLM推理 |
| PagedAttention | 非连续内存管理 | 50-80% | 页表开销 | 服务端LLM |
| GQA/MQA | 减少KV头数 | 按比例 | 精度损失 | 模型设计 |

#### 产业界最佳实践

1. **训练/微调**: `torch.utils.checkpoint` + 选择性检查点
2. **端侧推理**: 权重流式加载 + INT4量化 (llama.cpp/MLC-LLM)
3. **GPU服务端**: PagedAttention + 连续批处理 (vLLM)
4. **显存不足**: CPU Offloading + 异步预取 (DeepSpeed/Accelerate)
5. **LLM推理**: GQA + KV Cache量化 + 滑动窗口
6. **编译期优化**: TVM/MLC-LLM静态内存规划，零运行时开销
