# 17.2 在线学习本节涵盖：实时更新, 主动学习

```python
import torchimport torch.nn as nnimport torch.nn.functional as Ffrom collections import dequeimport randomtorch.manual_seed(42)class OnlineModel(nn.Module):    def __init__(self):        super().__init__()        self.net = nn.Sequential(nn.Linear(10, 32), nn.ReLU(), nn.Linear(32, 3))    def forward(self, x):        return self.net(x)model = OnlineModel()optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)replay_buffer = deque(maxlen=200)print('=== Online Learning with Replay Buffer ===')n_rounds = 20for round_idx in range(n_rounds):    X_new = torch.randn(8, 10) + round_idx * 0.2    y_new = torch.randint(0, 3, (8,))        replay_buffer.append((X_new, y_new))        loss = F.cross_entropy(model(X_new), y_new)        if len(replay_buffer) > 1:        X_old, y_old = random.choice(list(replay_buffer))        old_loss = F.cross_entropy(model(X_old), y_old)        loss = loss + 0.5 * old_loss        optimizer.zero_grad(); loss.backward(); optimizer.step()        if (round_idx + 1) % 5 == 0:        print(f'Round {round_idx+1}: loss={loss.item():.4f}, buffer_size={len(replay_buffer)}')print('\nReplay buffer prevents catastrophic forgetting by mixing old and new data.')print()print('=== Active Learning for Online Updates ===')print('1. Monitor model confidence on incoming data')print('2. Flag low-confidence samples for human review')print('3. Use reviewed samples for targeted updates')print('4. Prioritize samples near decision boundaries')print('5. Minimize annotation cost while maximizing model improvement')
```
