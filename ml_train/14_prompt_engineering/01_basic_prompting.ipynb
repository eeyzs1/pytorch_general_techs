# 14.1 基础提示本节涵盖：Zero-shot, Few-shot, Chain-of-Thought

```python
from transformers import AutoModelForCausalLM, AutoTokenizerimport torchtokenizer = AutoTokenizer.from_pretrained('gpt2')model = AutoModelForCausalLM.from_pretrained('gpt2')print('=== Zero-Shot Prompting ===')prompt_zero = 'Translate English to French: Hello'inputs = tokenizer(prompt_zero, return_tensors='pt')outputs = model.generate(**inputs, max_new_tokens=10, do_sample=False)print(f'Prompt: {prompt_zero}')print(f'Output: {tokenizer.decode(outputs[0])}')print()print('=== Few-Shot Prompting ===')prompt_few = '''Q: 2+2=? A: 4Q: 3+5=? A: 8Q: 7+1=? A:'''inputs = tokenizer(prompt_few, return_tensors='pt')outputs = model.generate(**inputs, max_new_tokens=5, do_sample=False)print(f'Prompt: {prompt_few}')print(f'Output: {tokenizer.decode(outputs[0])}')print()print('=== Chain-of-Thought (CoT) ===')prompt_cot = '''Q: Roger has 5 tennis balls. He buys 2 more cans of 3. How many does he have?A: Roger started with 5 balls. 2 cans of 3 = 6 balls. 5 + 6 = 11. The answer is 11.Q: The cafeteria had 23 apples. They used 20 for lunch and bought 6 more. How many?A:'''inputs = tokenizer(prompt_cot, return_tensors='pt')outputs = model.generate(**inputs, max_new_tokens=20, do_sample=False)print(f'Prompt: {prompt_cot}')print(f'Output: {tokenizer.decode(outputs[0])}')print()print('CoT dramatically improves reasoning on math and logic tasks.')
```
