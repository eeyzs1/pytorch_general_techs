# 11.2 查询处理本节涵盖：查询改写, 查询分解, HyDE, Step-back Prompting

```python
print('=== Query Rewriting ===')print('Original: "How to fix Python error?"')print('Rewritten: "How to debug and fix common Python runtime errors and exceptions?"')print('Adds context, removes ambiguity, expands synonyms')print()print('=== Query Decomposition ===')print('Complex: "Compare GPT-4 and Claude on coding, math, and reasoning"')print('Sub-queries:')print('  1. "GPT-4 coding benchmark performance"')print('  2. "Claude coding benchmark performance"')print('  3. "GPT-4 math benchmark performance"')print('  4. "Claude math benchmark performance"')print('  5. "GPT-4 reasoning benchmark performance"')print('  6. "Claude reasoning benchmark performance"')print()print('=== HyDE (Hypothetical Document Embeddings) ===')print('1. Generate hypothetical answer to query using LLM')print('2. Use hypothetical answer embedding to retrieve real documents')print('3. Real documents are closer to hypothetical answer than to query')print()print('Query: "What is quantum computing?"')print('Hypothetical answer: "Quantum computing uses qubits that can exist in superposition..."')print('Retrieve docs similar to this hypothetical answer')print()print('=== Step-back Prompting ===')print('Specific: "What was the US GDP growth rate in Q3 2023?"')print('Step-back: "What factors affect US GDP growth rates?"')print('Retrieve background knowledge, then answer specific question')
```
